```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis);
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis);
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis);
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis);
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend();
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the x limit between 0 and 45.
plt.xlim(0, 11)
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend();
```

```python
# Create the plot.
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2,label="Boston");
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Add the legend.
ax.legend()
# Set the x limit between 0 and 11.
ax.set_xlim(0, 11)
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# Add a grid.
ax.grid()
# Add title
ax.set_title("PyBer Fare by Month");
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c="red", label='Chicago')
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
plt.gca().invert_yaxis()
plt.xlim(0, 50)
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], c='skyblue', alpha=0.25,edgecolor="black",linewidth=3,label='Boston')
ax.legend()
ax.invert_yaxis()
ax.set_xlim(0, 50)
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("regression analysis");
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%');
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis,labels=x_axis,autopct='%.1f%%',explode=explode_values,shadow=True,startangle=95,counterclock=False,colors=colors)
plt.show()
```
